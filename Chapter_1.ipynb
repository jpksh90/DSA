The notion of computation predates the invention of computing machines. As is, a compute is a dumb machine, which can only perform fast computations with occasional storing the results. The way a computer is made intelligent is via providing a sequence on instructions to peforms a computation.

Otherwise, somebody can didcate the process to compute the double of a number in the form of a recipie. This is popularly known as imperative form, such as the recipie for comptuing the twice of an number.

```
1. Take a number, say, n
2. add 1 to the n for n times
3. Write the result
```

Choosing whether to use the declarative form or imperative form depdends on the multiple factors. Somebody with knowledge of multiplication can follow the first instruction. Anybody with only knowledge of addition will follow the second version but not the first one.

Though, this example seems trivial, however the implications are profound. These sort of understanding depends on the underlying "processor" on which the program runs. These set of instructions understood by a computer is called an algorithm. An algorithm usually shows the steps in plain text. Now, to make the algorithmn understood by the computer, one need to implement it in a programming language.

Programming language is a medium to express the set of computations to perform. Consider, programming language to be a tool. Like, tools vary in shape, size, and applications, so do, programming languages. We will come back to this topic in subsequent section.

## Section 1: Computation Model


Before delving into programming, let us reivew how does a computer work. The first component of a computer is a memory which can store data. The data stored is fetched and processed by a processor. When the processor has processed the data, it write the result to the memory.

### Von-Neumann Model

A computer has processor, a set of memory locations with input and output devices mapped to this memory location. This is an abstract model of computation, called as the Von Neumann Model, after its inventor, *John Von Neumann*.


You can visualize, a memory like a page with a number of square cells, arranged in a row and columns. Each column and row has an address. A processor reads and write to a specified memory cell address.

In addition, the program is also stored in the same memory, called as stored-program computer. On each read and write to the memory, the processor checks whether the address contains a instruction or a data. If it contains an instruction, it runs that instruction, otherwise, it fetces the data. You may argue, why isn't program stored at a separate memory location. In fact, it exists, popularly known as hand-held scientific calculators. Here, the program is written to a read-only memory, however, these are not flexible as they cannot be re-programmed. It will also not be optimal to keep a separate memory for the program.

Computers also take inputs from the keyboard or mouse and outputs to the screen. Where does these devices fit in? Here, the input and output devices are mapped to a fixed address in the memory, which is recognized by the device. The device also recognizes this address and reads/writes from/to this address.

### Turing Model

Von-Neuammn architecture is inspired from an abstract model of computer called Turing machine. Turing machine is a simple device, which consists of an infinite tape to read and write letters or symbols. It allows to move left and right in the tape. The turing machine stores the current state of the symbol and based on the input taken from the tape, it can either move to the left or right cell of the tape, shift to another state, or write to the memory.

Turing machines are important to computer science, as it demostrates the scope of problems solved by a computer. Any algorithm which is solvable can be expressed in a turing machine. We will stop here, and do not delve into the details of turing machine, as this is covered in the lecture on theoritical computer science.

# Binary Digits

Classical computers work on binary digits, which consists of the numbers 0 and 1. This notion was invented in the 1950s and has remained popular. Each binary digit is called a bit. A byte is a sequence of eight bits. This notion can be extended by multiplication factor, such as kilobyte, megabyte, gigabyte, and so on. Each factor multiplies the factor by 1024, as opposed to 1000 in other values.

| Count     | Unit     |
|-----------|----------|
| 1         | 1 Bit      |
| 8 bits      | 1 Byte     |
| 1024 Bytes | 1 Kilobyte |
| 1024 KBs   | 1 Megabyte |
| 1024 MBs  | 1 Gigabyte |
| 1024 GBs | 1 Terabyte |

Binary digit is motivated by the ease of implementation. Since, processor and memory are electrical devices, they work on the voltages. The prescence of voltage is considered as 1, while its absence is considered as 0.

Now, the question arises, how do you represent characters in binary digits. The answer is simple, map these characters to numbers. These mappings are called as encoding schemes. Two popular encoding schemes are the ASCII and UTF scheme. ASCII takes 1 byte of memory and can represent 256 characters. UTF takes 4 bytes of memory and can represent all languages and many other characters such as emojis.

### Decimal to Binary Conversion

Decimal to binary conversion proceeds by repeated integer division by 2. On each division, the remainder denotes the binary value at that particular location. For example, take a number 10.

1. 10 / 2 = 5, remainder = 0
2. 5 / 2 = 2, remainder = 1
3. 2 /2  = 1, remainder = 0
4. 1 / 2 = 0, remainder = 1

So the binary equivalent of 10 is 1010.

### Hexadecimal Numbers

Often it is inconvinent to write in Binary. Hexadecimal numbers are of base 16, where the $0-9$ maps to the decimal $0-9$ and $11-15$ maps to the $A-F$. Four binary digits form a hexadecimal number. Converting a decimal to hexadecimal is similar to binary, except that here the decimal number is divided by 16 instead of 2, and the remainder is map to the hexa-decimal code.

## Processor

A processor consists of the a few register memories, which have the least access time and are directly addressed by the processor. These registers are hard-coded and cannot be extended. There are a number of special purpose registers, such as instruction register and program counter. Instruction register stores the current instruction that is being executed. Program counter stores the index of the instruction which is next in the line of execution. A processor first checks the program counter, and then it loads the instruction at the corresponding memory to the instruction register to execute. This cycle is called as instruction-fetch-execute cycle.

Each processor has a set of binary coded instructions that it can understand. The wires in the processors are specifically wired to respond to these specific binary code. We call these codes as *instruction-set architecture*. Technically, these are set of instructions which are understood by each processor architecture. You can imagine it to your pet dog which can understand your instructions, however, fails to undertstands the instructions from your neighbour or friend. Nonetheless, these instructions are still in binary and it is difficult for human to remember all these binary combinations. Therefore, these are mapped to the short english instructions, which can be written in english alphabets. A special program called an assembler translates the assembly code to the binary coded instruction.
